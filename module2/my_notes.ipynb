One of the most common magic commands is the `%mathplotlib inline` command.  This command will enable plots to be displayed directly inside the notebook.

````python
# Display in the current window 
%matplotlib inline
````

````python
"""Simple demo of scatter plot"""
import numpy as np
import matplotlib.pyplot as plt

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
colors = np.random.rand(n)
# Calculate area
area = np.pi * (15*np.random.rand(n))**2 # 0 to 15 paint radius
#plot data
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
````

## Numpy

### Features
- High performance
- Vectorization
- Broadcasting
- Interoperability


### Comparison with other alternatives

#### Python List vs Numpy
- Elements of different types `vs` the same type
- Variable size nesting is possible `vs` same size
- No operations at the list level `vs` operations allowed at the array level
- Uses a lot of memory `vs` less memory
