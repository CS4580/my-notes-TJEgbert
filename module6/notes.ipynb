### Assess Seasonality
When meeting with the employees of the `Avila Adobe`, you hear the one major pain point is that the number visitors varies greatly with the season.
Part A:
- Create a `Line Chart` that shows how the number of visitors to the `Avila Adobe` has evolved over time
- Replace `inf` with `nan` and then fill them in with 0

```python
import numpy as np
# Replace infinite values with NaN
museum_df = museum_df.replace([np.inf, -np.inf], np.nan)
# Now we can use the fillna() methods to replace Nan with zero
museum_df = museum_df.fillna(0)
```

```python
# Plot the data
plt.figure(figsize=(16, 6))
sns.lineplot(data=museum_df['Avila Adobe'], label='Avila Adobe')
plt.title('Monthly Visitors to Avila Adobe')
```

## Scatter Plots

### Load the DataSet
We will use a (synthetic) dataset of medical insurance `data/insurance.csv`

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Set up complete')

# Load data
file_path = '../data/insurance.csv'
insurance_df = pd.read_csv(file_path)

# TOP record
insurance_df.head()
```

Create a `scatter` plot.  In seaborn use `sns.scatterplot` command.
- x-axis use BMI
- y-axis use charges


```python
sns.scatterplot(x=insurance_df.bmi, y=insurance_df.charges)
```

The plot suggest we have a `positive correlation` between BMI and charges.
- Higher BMI tends to pay more

To double check the strength of this relationship, add a `regression line` that best fits the data.

```python
sns.regplot(x=insurance_df.bmi, y=insurance_df.charges)
```

### Color-coded Scatter plots
They are good when you need to display the correlation between not two, but more variables.

We want to understand how `smokers` affects the relationship between bmi and `charges`.

```python
# Scatter plot of bmi vs charges AND smokers vs charges
# x = bmi, y = charges, hue = smoker.
sns.scatterplot(x= insurance_df.bmi, y=insurance_df.charges, hue=insurance_df.smoker)
```

```python
# Add two regression lines
# When using lmplot for more three of more, only use the column name
sns.lmplot(data=insurance_df,x= 'bmi', y='charges', hue='smoker')
```

Finally, we can adapt the design of a scatter plot to feature `categorical variable` (like `smoker`) on one of the main axis.  Some people call this plot **Categorical Scatter Plot**

```python
sns.swarmplot(x=insurance_df.smoker, y=insurance_df.charges)
```
