## Introduction to Plotting

### Python Packages for plotting
- `matplotlib`: All plots in pandas are matplotlib objects
- `seaborn`: Focus on statistical visualization
- `bokeh`: To create interactive visualization for the internet
- `Plotly`: Create interactive embedded visualization


### Matplotlib
Main submodule is `pyplot` which will give you different function plot elements

### Creating a Plot
- Import the necessary packages
- Import or create the data
- Call the appropriate plotting functions

```python
import pandas as pd
import matplotlib.pyplot as plt
# Display the plots inside the notebook
%matplotlib inline
```

```python
# Dummy data
x = [1, 2, 3, 4, 5]
y = [12, 23, 34, 45, 89]

print(x)
print(y)
```

```python
# Default is line plot
plt.plot(x,y)
#plt.show()
```
