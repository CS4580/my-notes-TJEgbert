### Bar Plot
Useful when you are showing or comparing segments of information or categorical data

```python
cities_df = pd.read_csv('../data/cities.csv')
cities_df.head()
```

```python
# Filter records from Year = 2019
cities_2019_df = cities_df[cities_df['Year'] == 2019]
cities_2019_df.head()
```

```python
# Now plot a bar chart of the numbers of cities in each region
cities_2019_df.plot(kind='bar', x='City', y='Population', title='Population of Cities in 2019', figsize=(10, 6))
```

```python
# You may want to plot data horizontally
# Since we are only plotting one set of values, we do not need to legend
cities_2019_df.plot(kind='barh', x='City', y='Population', title='Population of Cities in 2019', legend=False, figsize=(10, 6))
```

### Scatter Plots
When you are interested in the relationship between two variables, scatter plots are probably the best choice

```python
# Load data
houses_df = pd.read_csv('../data/homes.csv')
houses_df.head()
```

```python
# Plot the scatter plot
houses_df.plot(kind='scatter', x='List', y='Rooms')
```

```python
# Filter data directly
houses_df[houses_df['Age'] < 50].plot(kind='scatter', x='Age', y='List')
```

```python
# If you want to eliminate outliers, you can filter and data before plotting
houses_df_plot = houses_df[houses_df['Age'] < 50]
houses_df_plot = houses_df_plot[houses_df_plot['List'] < 350]
# Now plot
houses_df_plot.plot(kind='scatter', x='Age', y='List', title='House that are less than 50 years old and list is less then 350')
```

### Histograms
Use to represent frequency in data

```python
houses_df_plot.head()
```

```python
# If you want to know the distribution of acreage for the houses
# sold in filtered dataset, use a histogram
houses_df_plot['Acres'].plot(kind='hist', title='Acre Distribution for home < 50 and list < 350')
```

### Area Plot
Useful when you want to understand how values develope and are composed over time

```python
# Load data set
enrollments_df = pd.read_csv('../data/enrollment.csv')
enrollments_df.head()
```

```python
# Create an area plot of the enrollments
enrollments_df.plot(kind='area', x='Year')
```

```python
# Data is not good because it is not organized by year
# We want to see the total, and how it changes over time
# Sort the data by year, then plot it using the pivot method
pivot_df = enrollments_df.pivot(index='Year', columns='Class', values='Enrollment')
pivot_df.head()
```

```python
# plot it
pivot_df.plot(kind='area')
```

```python
# The data is staked by default.
# You can set this option to false, stack=False
pivot_df.plot(kind='area', stacked=False)
```
